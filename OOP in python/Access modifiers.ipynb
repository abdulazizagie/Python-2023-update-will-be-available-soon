## Access Specifiers/Modifiers

Access specifiers or access modifiers in python programming are used to limit the access of class variable and class methods outside of class while implementing the concepts of inheritance.

Types of a acces specifiers
1. Public access modifier
2. private access modifier
3. Protected access modifier

## Public Access Specifier in python
All the variable and method (member funcation) in python are by defult public ,Any instance variable in a class followed bt the'self' keyword ie. self.var_name are public accessed.

### Example:
```python
class Student:
    def __init__(self,age,name):
        self.age = age   # public variable
        self.name = name  # public variable

obj = student(21,"Harray")
print(obj.age)
print(obj.name)
```
#### Output
```python
21
Harray
```

## Private Access Modifier

By definition , privite member of a class (variable or methods) are those some other programming language.However, a convention has been established to indicate that a variable or method should be considered private by profixing its name with an underscore(_).This is know as a "weak internal use indicator" and it is a convention only, not a strict rule. Code outside the class can still access these"private" variable and methods,but it is generally understood that they should not be accessed or modified.

### Example:

### Name mangling
Name maangling in python is a technique used to protect class-private and superclass-private attributes from being accidentally overwritten by subclass.Name of class-private and superclass-private attributes are transformed by the addition of a single leading underscore and a double leading underscore respectively.
```python
class MyCalss:
    def __init__(self):
        self.__private_attribute = "I am a private attribute"
        self.__mangled_attribute = "I am a mangled  attribut"

my_object = Myclass()

print(my_object._private_attribute) # Output: I am a privite attribute

print(my_object.__mangled_attribute) # Throws an AttributeError

print(my_object._Myclass__mangled_attribute) # Output: I am a mangled attribute
```

In the example above, the attribute_private is marked as private by convention,but can still be accessed from outside the class. The attribute __mangle_attrivude is private and its name is "mangled" to outside the class,but you can access it by calling_Myclass__mangled_attribute

It's important to note that the single underscore is just a naming convention , and does not actuall provide any protection or restrict access to the member .The syntax we follow to amke any variable project is to write variable name followed by a single underscore(_) ie._variable.
